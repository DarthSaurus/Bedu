```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Cargar Dataset
data = pd.read_csv('/content/drive/MyDrive/Remoto Datasets/Remoto melbourne_housing-clean.csv')

#Mostrar la información basica
data_info = data.info()
data_head = data.head()
data_describe = data.describe()
```

Para este ejercicio vamos a utilizar el dataset 'melbourne_housing-clean.csv'. La fuente original [está aquí](https://www.kaggle.com/anthonypino/melbourne-housing-market), en caso de que quieras recordar de qué trata.

Vamos a realizar 3 gráficas de barras utilizando este dataset. Las instrucciones para cada una son las siguientes:

1. Gráfica de barras *verticales* usando la columna `rooms`. El eje y será el *conteo de frecuencia* de cada categoría. Las barras tienen que quedar *ordenadas en orden descendente* (es decir, de la más larga a la màs corta). Para lograr ordenarlas, tendrás que revisar [la documentación](https://seaborn.pydata.org/generated/seaborn.barplot.html) de la gráfica de barras de Seaborn.

2. Gráfica de barras *verticales* de la columna `types`. El eje y será el *porcentaje de frecuencia* de cada categoría.

3. Gráfica de barras *h*orizontales* de la columna `council_area`. El eje x será el *conteo de frecuencia* de cada categoría.

**Todas** las gráficas deberán de tener títulos y nombres adecuados para los ejes. También deberán de tener el tamaño correcto para que la visualización sea comprensible.

```python
#Grafica 1

# Configurar estilo de gráficos
sns.set_style("whitegrid")

# Crear figura y ejes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Crear gráfica de barras con Seaborn
sns.countplot(data=data, x='rooms', ax=ax1, order=data['rooms'].value_counts().index)

# Títulos y etiquetas de ejes
ax1.set_title('Frecuencia de Cantidad de Habitaciones', fontsize=16)
ax1.set_xlabel('Número de Habitaciones', fontsize=12)
ax1.set_ylabel('Frecuencia', fontsize=12)

# Mostrar gráfica
plt.show()
```

```python
# Grafica 2

# Crear figura y ejes
fig, ax2 = plt.subplots(figsize=(10, 6))

# Calcular porcentajes y crear gráfica de barras con Pandas
(type_counts := data['type'].value_counts(normalize=True)*100).plot(kind='bar', ax=ax2, color=sns.color_palette("viridis", len(type_counts)))

# Títulos y etiquetas de ejes
ax2.set_title('Porcentaje de Frecuencia de Tipos de Propiedades', fontsize=16)
ax2.set_xlabel('Tipo de Propiedad', fontsize=12)
ax2.set_ylabel('Porcentaje (%)', fontsize=12)

# Mostrar gráfica
plt.show()
```

```python
# Grafica 3

# Crear figura y ejes
fig, ax3 = plt.subplots(figsize=(10, 6))

# Crear gráfica de barras horizontal con Seaborn
sns.countplot(data=data, y='council_area', ax=ax3, order=data['council_area'].value_counts().index, palette="viridis")

# Títulos y etiquetas de ejes
ax3.set_title('Frecuencia de Áreas de Consejo Local', fontsize=16)
ax3.set_xlabel('Frecuencia', fontsize=12)
ax3.set_ylabel('Área del Consejo Local', fontsize=12)

# Mostrar gráfica
plt.show()
```
