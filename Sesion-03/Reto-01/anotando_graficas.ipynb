```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Remoto Datasets/Remoto diabetes-clean.csv', index_col=0)

df.head()
```

### a)

```python
# crear la figura matlob
plt.figure(figsize=(12, 6))

# Crear un  histograma de la columna de la glucosa usando histplot
sns.histplot(df['glucose'], kde=False, bins=30)

# Agregar etiquetas y titulo
plt.title('Distribution of Glucose Levels', fontsize=16)
plt.xlabel('Glucose Level', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Mostrar la grafica
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### b)

```python
pregnancy_counts = df['pregnancies'].value_counts()
```

```python
# crear la figura matlob
plt.figure(figsize=(12, 6))

sns.barplot( x = pregnancy_counts.index, y= pregnancy_counts)

# Agregar etiquetas y titulo
plt.title('Distribution of Number of Pregnancies', fontsize=16)
plt.xlabel('Number of Pregnancies', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Mostrar la grafica
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### c)

```python
#Deividir el dataframe entre individuos saludables y enfermos
healthy = df[df['outcome'] == 0]
sick = df[df['outcome'] == 1]
```

```python
# crear la figura matlob
plt.figure(figsize=(12, 6))

# Crear una grafica de frecuencias de la glucosa en individuos saludables
sns.distplot(healthy['glucose'], hist=False, kde_kws={'shade': True}, label='Saludable', color='green')
# Crear una grafica de frecuencias de la glucosa en individuos enfermos
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True}, label='Enfermo', color='red')

#Agregamos Etiquetas y Nombres
plt.title('Frecuencia de los Niveles de Glucosa en individuos saludables y enfermos', fontsize=16, pad=10)
plt.xlabel('Nivel de Glucosa', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.legend(title='Individuos', title_fontsize='13', fontsize='12')
```

### d)

```python
# crear la figura matlob
plt.figure(figsize=(12, 6))

# Crear una grafica de frecuencias de los niveles de Presión Sanguinea, Glucosa e Insulina
sns.distplot(sick['blood_pressure'], hist=False, kde_kws={'shade': True}, label='Presion Sanguinea')
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True}, label='Glucosa')
sns.distplot(sick['insulin'], hist=False, kde_kws={'shade': True}, label='Insulina')

#Agregamos Etiquetas y Nombres
plt.title('Frecuencia de los Niveles de Presión Sanguinea, Glucosa e Insulina en individuos enfermos', fontsize=16, pad=10)
plt.xlabel('Niveles de Presión Sanguinea, Glucosa e Insulina', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.legend(title='Frecuencias', title_fontsize='13', fontsize='12')
```
