```python
import pandas as pd

df = pd.read_csv('../../Datasets/near_earth_objects-jan_feb_1995-clean.csv', index_col=0)
diameter_column = 'estimated_diameter.meters.estimated_diameter_max'
total_count = df.shape[0]
mean = df[diameter_column].mean()
std = df[diameter_column].std()

within_one_std_filter_bottom = df[diameter_column] >= (mean - std)
within_one_std_filter_top = df[diameter_column] <= (mean + std)
within_one_std = df[within_one_std_filter_bottom & within_one_std_filter_top]
percentage_of_data_within_one_std = within_one_std.shape[0] * 100 / total_count

within_two_std_filter_bottom = df[diameter_column] >= (mean - 2 * std)
within_two_std_filter_top = df[diameter_column] <= (mean + 2 * std)
within_two_std = df[within_two_std_filter_bottom & within_two_std_filter_top]
percentage_of_data_within_two_std = within_two_std.shape[0] * 100 / total_count

within_three_std_filter_bottom = df[diameter_column] >= (mean - 3 * std)
within_three_std_filter_top = df[diameter_column] <= (mean + 3 * std)
within_three_std = df[within_three_std_filter_bottom & within_three_std_filter_top]
percentage_of_data_within_three_std = within_three_std.shape[0] * 100 / total_count
```

````python
# 4. Percentage of samples within 1 standard deviation from the mean
mean_value = df_meteorites['estimated_diameter.meters.estimated_diameter_max'].mean()
samples_within_1std = df_meteorites[(df_meteorites['estimated_diameter.meters.estimated_diameter_max'] >= mean_value - std_dev) &
                                   (df_meteorites['estimated_diameter.meters.estimated_diameter_max'] <= mean_value + std_dev)]
percentage_1std = (len(samples_within_1std) / total_samples) * 100

# 5. Percentage of samples within 2 standard deviations from the mean
samples_within_2std = df_meteorites[(df_meteorites['estimated_diameter.meters.estimated_diameter_max'] >= mean_value - 2*std_dev) &
                                   (df_meteorites['estimated_diameter.meters.estimated_diameter_max'] <= mean_value + 2*std_dev)]
percentage_2std = (len(samples_within_2std) / total_samples) * 100

# 6. Percentage of samples within 3 standard deviations from the mean
samples_within_3std = df_meteorites[(df_meteorites['estimated_diameter.meters.estimated_diameter_max'] >= mean_value - 3*std_dev) &
                                   (df_meteorites['estimated_diameter.meters.estimated_diameter_max'] <= mean_value + 3*std_dev)]
percentage_3std = (len(samples_within_3std) / total_samples) * 100

percentage_1std, percentage_2std, percentage_3std
````
